```python
import pandas as pd
from sqlalchemy import create_engine
import psycopg2 
import sys
```

```python
# wine_review_file = "winemag-data-130k-v2.csv"
Wine_df = pd.read_csv("winemag-data-130k-v2.csv")
Wine_df
```

```python
 Wine_df.info()
```

```python
Wine_df = Wine_df[['country','description','designation','points','price','region_1','variety','winery']]
```

```python
# Wine_df["Unnamed"] = Wine_df["Unnamed"].fillna(0)
Wine_df["country"] = Wine_df["country"].fillna("Unknown")
Wine_df["description"] = Wine_df["description"].fillna("Unknown")
Wine_df["designation"] = Wine_df["designation"].fillna("Unknown")
Wine_df["points"] = Wine_df["points"].fillna(0)
Wine_df["price"] = Wine_df["price"].fillna(0)
#Wine_df["province"] = Wine_df["province"].fillna("Unknown")
Wine_df["region_1"] = Wine_df["region_1"].fillna("Unknown")
#Wine_df["region_2"] = Wine_df["region_2"].fillna("Unknown")
# Wine_df["taster_name"] = Wine_df["taster_name"].fillna("Unknown")
# Wine_df["taster_twitter_handle"] = Wine_df["taster_twitter_handle"].fillna("Unknown")
# Wine_df["title"] = Wine_df["title"].fillna("Unknown")
Wine_df["variety"] = Wine_df["variety"].fillna("Unknown")
```
