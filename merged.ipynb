```python
import pandas as pd
from sqlalchemy import create_engine
import psycopg2 
# import sys
```

```python
csv_file = "wine-reviews/winemag-data_first150k.csv"
wine_data_df = pd.read_csv(csv_file)
wine_data_df
```

```python
wine_data_df.info()
```

```python
wine_data_df = wine_data_df[['country', 'description', 'designation', 'points','price', 'region_1','variety','winery']]
```

```python
wine_data_df['description'] = wine_data_df['description'].fillna('Unknown')
wine_data_df['designation'] = wine_data_df['designation'].fillna('Unknown')
wine_data_df['price'] = wine_data_df['price'].fillna(0)
#wine_data_df['province'] = wine_data_df['province'].fillna('Unknown')
wine_data_df['region_1'] = wine_data_df['region_1'].fillna('Unknown')
wine_data_df['variety'] = wine_data_df['variety'].fillna('Unknown')
wine_data_df['winery'] = wine_data_df['winery'].fillna('Unknown')
wine_data_df['country'] = wine_data_df['country'].fillna('Unknown')
```

```python
wine_data_df.info()
```

```python
engine = create_engine('postgresql://postgres:ALberto!987@localhost:5432/wine_db')
```

```python
engine.table_names()
```

```python
wine_data_df.to_sql(name='wine_reviews', con=engine, if_exists='append', index=False)
```
